# 1.  Clean Dataset

```python
# Transforming csv object to Data Frame
games = pd.read_csv('vgsales.csv')
# Creating copy 
games_copy = games.copy()
```

```python
games_copy.head()
```

```python
games_copy.info(memory_usage='deep')
```

```python
games_copy.isnull().sum()
```

```python
# deleting NA values from Year column
games_copy.dropna(subset=['Year'],inplace=True)
```

```python
games_copy.isnull().sum()
```

```python
# checking NaN values for publisher columns
games_copy.loc[games_copy.Publisher.isnull()]
```

```python
# changing Nan values
games_copy.Publisher.fillna('Unknown Publisher', inplace=True)
```

```python
# changing index to title of the game
games_copy.set_index(games_copy.Name, inplace=True)
```

```python
games_copy.head()
```

```python
# Checking for duplicates in Data Frame
games_copy.loc[games_copy.duplicated()]
```

```python
games_copy.describe()
```

```python
games_copy.info(memory_usage='deep')
```
